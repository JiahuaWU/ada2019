```python
def match_count_ing(ing_list,ingredient):
    match = 0
    for ings in ing_list.split(', '):
        for ing in ingredient:
            if ing==ings.replace('\'','').replace('[','').replace(']',''):
                match+=1
            elif ing in ings:
                    match+=0.5
    return match
```

```python
df[(df.ingredients.apply(match_count_ing,args=(ingredient,))/df.n_ingredients)>0.7]
```

```python
name_hist = input().split(',')
```

```python
name_str = " ".join(name_hist).replace('\'','').replace('[','').replace(']','')
```

```python
def match_count_name(name,name_str):
    match = 0
    for n in name.split():
        for ns in name_str.split():
            if n==ns:
                match+=1
    return match/len(name_str.split())
```

```python
df.dropna(inplace=True)
```

```python
df[df.name.apply(match_count_name,args=(name_str,))>0.3]
```

```python
match_count_name(df.name.iloc[0],name_str)
```
